```python
# here tips and flights are the built-in datasets in seaborn
```

     Heat Map

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips_corr = tips[['total_bill', 'tip']]
```

```python
tips_corr
```

```python
b = tips_corr.corr()
b
```

```python
sns.heatmap(b, annot = True )
```

    Cluster Map

```python
sns.clustermap(b)
```

```python
import pandas as pd
flights.head()
```

```python
sns.load_dataset("flights")
```

```python
pivot = flights.pivot_table(values = 'passengers', index = 'month', columns= 'year')
```

```python
pivot
```

```python
pivot.plot(kind='bar')
plt.title('Average Tip by Day and Sex')
plt.ylabel('Average Tip')
plt.xlabel('Day')
plt.legend(title='Sex')
plt.show()
```
