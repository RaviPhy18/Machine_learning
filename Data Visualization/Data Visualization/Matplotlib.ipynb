```python
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.plot(x**2,y**25)
plt.subplot(2,2,3)
plt.plot(x*50,y*(1/9))
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

Object Oriented way

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch)

```python
plt.figure()
```

```python
fig = plt.figure()
```

```python
fig = plt.figure(figsize = (12,8), dpi= 100)
axis1 = fig.add_axes([0.1,0.1,0.4,.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.6,0.5,0.5])
axis2.plot(x,y)
```
